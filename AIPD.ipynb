```python
y_pred = LR.predict(x_test)
```

```python
print("Accuracy : ", ACC)
print("\nClassification Report : \n", CR)
```

```python
SVM.fit(x_train, y_train)
```

```python
RF.fit(x_train, y_train)
```

```python
y_pred2 = RF.predict(x_test)
```

```python
print("Accuracy : ", ACC2)
print("\nClassification Report : \n", CR2)
```

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(32, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=20, batch_size=32)
```

```python
selection_rate = df.groupby('gender')['shortlisted'].mean()
print(selection_rate)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()

LR.fit(x_train, y_train)
y_pred_LR = LR.predict(x_test)

from sklearn.metrics import accuracy_score, classification_report
ACC_LR = accuracy_score(y_test, y_pred_LR)
CR_LR = classification_report(y_test, y_pred_LR)

print("Accuracy : ", ACC_LR)
print("\nClassification Report : \n", CR_LR)
```

```python
from sklearn.svm import SVC
SVM = SVC(class_weight='balanced')

SVM.fit(x_train, y_train)
y_pred_SVM = SVM.predict(x_test)

from sklearn.metrics import accuracy_score, classification_report
ACC_SVM = accuracy_score(y_test, y_pred_SVM)
CR_SVM = classification_report(y_test, y_pred_SVM)

print("Accuracy : ", ACC_SVM)
print("\nClassification Report : \n", CR_SVM)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train / 255.0
y_train = y_train / 255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3)
```

```python
loss, acc = model.evaluate(x_test, y_test)
print("Test Accuracy : ", acc)
```

```python
image = x_test[3]
label = y_test[3]

noise = np.random.normal(0, 0.1, image.shape)
adversarial_image = image + noise
adversarial_image = np.clip(adversarial_image, 0, 1)
```

```python
org_pred = model.predict(image.reshape(1,28,28))
adv_pred = model.predict(adversarial_image.reshape(1,28,28))

print("Original Prediction : ", np.argmax(org_pred))
print("Adversarial Prediction : ", np.argmax(adv_pred))
```

```python
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.title("Adversarial Image")
plt.imshow(adversarial_image)

plt.plot()
```
